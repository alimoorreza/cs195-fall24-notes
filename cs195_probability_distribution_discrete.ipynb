```python
import matplotlib.pyplot as plt
```

##Letâ€™s assume X is random variable which can take discrete values
> P(X) is a function that maps from all possible values of X to the probability of the corresponding event

> For example, $X$ is a random variable which can take one of the 2 values from {TRUE, FALSE}


```python
prob_x = [0.75, 0.5]

plt.figure(figsize=(4,2))
plt.bar(['Head', 'Tail'], prob_x, color=['blue', 'green'])
plt.ylim(0, 1)
plt.title('Probability distribution of biased coin toss')
plt.ylabel('Probability of X')
plt.xlabel('Discrete random variable X')
plt.show()
```

> Another example, X is a random variable which can take one of the 6 values for a standard six-sided die {1, 2, 3, 4, 5, 6}


```python
prob_x = [0.3, 0.3, 0.1, 0.1, 0.1, 0.1]

plt.figure(figsize=(4,2))
plt.bar(['1', '2', '3', '4', '5', '6'], prob_x, color=['red', 'blue', 'green', 'cyan', 'orange', 'yellow'])
plt.ylim(0, 1)
plt.title('Probability Distribution of a standard six-sided die')
plt.ylabel('Probability of X')
plt.xlabel('Discrete random variable X')
plt.show()
```
